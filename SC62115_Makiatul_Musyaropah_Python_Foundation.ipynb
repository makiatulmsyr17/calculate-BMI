2. Calculate the BMI of each student with the formula BMI=Weight (kg)/(Height (m))^2
3. Create a program to calculate BMI!
4. Update the df dataframe by adding the calculation result of question number 2, name the BMI column.


```python
# Hitung BMI dan tambahkan kolom 'BMI' ke dalam DataFrame
df_21['BMI'] = df_21["Berat badan (kg)"] / ((df_21["Tinggi Badan (cm)"] / 100) ** 2)

# Menampilkan DataFrame hasil
display(df_21)
```

5. Create a BMI status for each student from the following categories, then save the status data with the column name BMI Status.
Normal: 18.5-24.9
Overweight: 25.0-29.9
Obesity: 30.0-34.9
Underweight: < 18.5


```python
# Tentukan fungsi untuk mengkategorikan BMI ke dalam status
def categorize_bmi(bmi):
    if bmi < 18.5:
        return 'Underweight'
    elif 18.5 <= bmi <= 24.9:
        return 'Normal'
    elif 25.0 <= bmi <= 29.9:
        return 'Overweight'
    elif 30.0 <= bmi <= 34.9:
        return 'Obesity'
    else:
        return 'Unknown'

# Buat kolom 'Status BMI' berdasarkan kategori BMI
df_21['Status BMI'] = df_21['BMI'].apply(categorize_bmi)

# Menampilkan DataFrame hasil
display(df_21)
```

Perusahaan Investasi Jamin-Sugih Co kami memberikan penawaran investasi terbaik dengan pembagian keuntungan tetap sebesar 25% setiap tahun.

Buatlah program Python yang dapat menerima input dari pengguna dalam bentuk:
Modal Investasi (bilangan bulat antara 0-1.000.000.000)
Jangka Waktu Investasi (bilangan bulat antara 1-25 dalam tahun)
Selain itu, program akan menampilkan error message bahwa modal atau tenure tidak sesuai ketentuan.

Program akan menampilkan informasi akumulasi investasi setiap tahun, menggunakan iterasi untuk setiap tahun.

Contoh:
Input
>> Money Capital : 100000
>> Investment Tenure (years) : 2
Output
>> Year 1, your total investment : 125000.0
>> Year 2, your total investment : 156250.0


```python
def hitung_investasi():
  try:
    # Mendapatkan input modal dan jangka waktu
    modal = int(input("Money Capital (bilangan bulat antara 0-1.000.000.000): "))
    jangka_waktu = int(input("Investment Tenure (years) (bilangan bulat antara 1-25 dalam tahun): "))

    # Validasi input
    if modal < 0 or modal > 1000000000:
      raise ValueError("Modal tidak boleh kurang dari 0 atau lebih dari 1.000.000.000")
    if jangka_waktu < 1 or jangka_waktu > 25:
      raise ValueError("Jangka waktu tidak boleh kurang dari 1 atau lebih dari 25 tahun")

    # Menghitung keuntungan dan total investasi setiap tahun
    for tahun in range(1, jangka_waktu + 1):
      keuntungan = modal * 0.25
      total_investasi = modal + keuntungan * tahun
      print(f" Year {tahun}, your total investment: {total_investasi:.2f}")

  except ValueError as err:
    print(f"Error: {err}")

# Menjalankan fungsi
hitung_investasi()
```
